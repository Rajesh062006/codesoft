```python


print("===== Student Chatbot =====")
print("Type 'exit' to quit.\n")

while True:
    user = input("You: ").lower().strip()

    if user == "admission":
        print("Bot: Admissions are open. Visit the admission office for details.")

    elif user == "courses":
        print("Bot: We offer BCA, B.Sc, B.Com, BBA, and MCA courses.")

    elif user == "fees":
        print("Bot: Please contact the accounts office for the latest fee structure.")

    elif user == "library":
        print("Bot: The library is open from 9:00 AM to 5:00 PM.")

    elif user == "hostel":
        print("Bot: Hostel facilities are available for both boys and girls.")

    elif user == "contact":
        print("Bot: You can contact the college office at +91-9876543210.")

    elif user == "principal":
        print("Bot: The principal is available during office hours.")

    elif user == "thank you":
        print("Bot: You're welcome! Happy to help.")

    elif user == "exit":
        print("Bot: Thank you for using Student Chatbot. Goodbye!")
        break

    else:
        print("Bot: Sorry, I don't have information about that.")
```
